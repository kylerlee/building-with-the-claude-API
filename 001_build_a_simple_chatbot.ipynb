```python
from anthropic import Anthropic

client =  Anthropic()
model = "claude-sonnet-4-0"
```

```python
def add_user_message(messages, text):
    user_message = {
        "role": "user",
        "content": text
    }
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {
        "role": "assistant",
        "content": text
    }
    messages.append(assistant_message)

def chat(messages):
    message_response = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages
    )
    add_assistant_message(messages, message_response.content[0].text)
    print("claude >", message_response.content[0].text)
```

```python
# Build a simple chatbot
cached_messages = []
isSessionContinue = True

while isSessionContinue:
    user_input = input("> ")
    if(len(user_input) > 0):
        print("user >", user_input)
        add_user_message(cached_messages, user_input)
        chat(cached_messages)
    else:
        print("user >", "empty entry, exit session")
        isSessionContinue = False
```
