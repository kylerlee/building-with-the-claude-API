```python
%pip install anthropic python-dotenv

from dotenv import load_dotenv
from anthropic import Anthropic

client =  Anthropic()
model = "claude-sonnet-4-0"

def add_user_message(messages, text):
    user_message = {
        "role": "user",
        "content": text
    }
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {
        "role": "assistant",
        "content": text
    }
    messages.append(assistant_message)

def chat(messages, system=None, temperature=0.5):
    message_params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature,
    }

    if(system != None):
        message_params["system"] = system
    
    message_response = client.messages.create(**message_params)
    add_assistant_message(messages, message_response.content[0].text)
    print("claude >", message_response.content[0].text)
```
