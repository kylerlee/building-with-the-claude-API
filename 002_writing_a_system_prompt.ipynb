```python
%pip install anthropic python-dotenv

from dotenv import load_dotenv
from anthropic import Anthropic

client =  Anthropic()
model = "claude-sonnet-4-0"

def add_user_message(messages, text):
    user_message = {
        "role": "user",
        "content": text
    }
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {
        "role": "assistant",
        "content": text
    }
    messages.append(assistant_message)

def chat(messages, system=None):
    message_params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages
    }

    if(system != None):
        message_params["system"] = system
    
    message_response = client.messages.create(**message_params)
    add_assistant_message(messages, message_response.content[0].text)
    print("claude >", message_response.content[0].text)
```

```python
# Build a concise python function generator

cached_messages = []

add_user_message(
    cached_messages,
    "Write a Python function that checks a string for duplicate characters."
)

system_prompt = """
You are a experienced Python engineer who writes very concise code that is straight forward and easy to understand.
Keep the code responses as short as possible.
"""

chat(messages=cached_messages, system=system_prompt)
```
